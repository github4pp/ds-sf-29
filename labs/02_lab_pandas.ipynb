```python
# 6. Create a scatter plot between spirit servings and wine servings of all countries
drinks.plot(kind='scatter', x='spirit_servings', y='wine_servings',title='Drinking Scatter Plot',legend=True,grid=True,color='Red')
```

```python
# 7. Show a list of the top 10 spirit drinking countries 
# (show only country names and spirit servings)
desc_country = drinks.sort_values(by='wine_servings',ascending=False,)
desc_country[:10]
```

```python
# 8. Plot 6 histograms of wine servings by continent, 
# remember to share x and share y axis scales!
drinks.wine_servings.hist(by=drinks.continent,figsize=(8,8))
plt.show()
```

```python
# 9. What is the average wine consumption in South America?
df_sa = drinks[drinks.continent == 'SA']
df_sa['wine_servings'].mean()
```

```python
# 10. Which continent has the highest on average wine consumption?
#cont = drinks.groupby('continent').aggregate(sum)
cont = drinks.groupby('continent').mean()
contsort = cont.sort_values(by='wine_servings',ascending=False)
contsort['wine_servings'][:1]
```
