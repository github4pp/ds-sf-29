```python
import pandas as pd
from pandas.io import sql
```

Pandas can be used to connect to most relational databases. In this demonstration, we will create and connect to a SQLite database. SQLite creates portable SQL databases saved in a single file. These databases are stored in a very efficient manner and allow fast querying, making them ideal for small databases or databases that need to be moved across machines.

If you are looking to start using a database without the setup of `mysql` or `postgres`, SQLite is a good start.

```python
import sqlite3
```

```python
conn = sqlite3.connect('dat-test.db')
```

Let's return to the Rossmann sales data and load that into the database.

Data is moved to the database through the `to_sql` command, similar to the `to_csv` command.

`to_sql` takes as arugments:
    - `name`, the table name to create
    - `con`, a connection to a database
    - `index`, whether to input the index column
    - `schema`, if we want to write a custom schema for the new table
    - `if_exists`, what to do if the table already exists. We can overwrite it, add to it, or fail

```python
data = pd.read_csv('../../../lesson-15/assets/dataset/rossmann.csv', low_memory=False)
data.head()
```

```python
data.to_sql('rossmann_sales',
            con=conn,
            if_exists='replace',
            index=False)
```

```python
sql.read_sql('select * from rossmann_sales limit 10', con=conn)
```

```python
#Load the Walmart sales data
data1 = pd.read_csv('../../../lesson-16/assets/dataset/walmart-sales.csv', low_memory=False)
data1.head()
```

```python
#Create a table for walmart dataset
data1.to_sql('walmart',
            con=conn,
            if_exists='replace',
            index=False)
```

```python
#Load the features data
data2 = pd.read_csv('../../../lesson-16/assets/dataset/features.csv', low_memory=False)
data2.head()
```

```python
#Create a table for features dataset
data2.to_sql('feature',
            con=conn,
            if_exists='replace',
            index=False)
```

```python
#Select the store, date and fuel price on days it was over 90 degrees

sql.read_sql('select walmart.Store,walmart.date,feature.Fuel_Price from walmart,feature where walmart.Store = feature.Store AND feature.Temperature>90 limit 10', con=conn)
```

```python
#Select the store, date and weekly sales and temperature
sql.read_sql('select walmart.Store,walmart.date,walmart.Weekly_Sales,feature.Temperature from walmart,feature where walmart.Store = feature.Store limit 10', con=conn)
```

```python
q = '''select 
       walmart.Store,walmart.date,walmart.Weekly_Sales,feature.Temperature 
       from 
       walmart,feature where walmart.Store = feature.Store limit 10'''
sql.read_sql(q, con=conn)
```

```python
#What were average sales on holiday vs. non-holiday sales
q = '''select IsHoliday, avg(Weekly_Sales) from walmart GROUP BY IsHoliday'''
sql.read_sql(q, con=conn)
```

```python
#What were average sales on holiday vs. non-holiday sales when the temperature was below 32 degrees
q = '''SELECT 
        walmart.IsHoliday, avg(Weekly_Sales) 
            from walmart JOIN feature ON walmart.Store = feature.Store
            where feature.Temperature<32
            GROUP BY walmart.IsHoliday'''
sql.read_sql(q, con=conn)
```
